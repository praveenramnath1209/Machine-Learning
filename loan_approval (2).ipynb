```python


df = pd.read_csv("loan_approval_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder



le = LabelEncoder()
df[' education'] = le.fit_transform(df[' education'])
df[' self_employed'] = le.fit_transform(df[' self_employed'])
df[' loan_status'] = le.fit_transform(df[' loan_status'])  # 1: Approved, 0: Rejected
```

```python
from sklearn.linear_model import LogisticRegression
# Redundant LabelEncoders removed as encoding is done in the previous cell
lr=LogisticRegression()

# Use the corrected column names without leading spaces
i=df[[" no_of_dependents"," education"," self_employed"," income_annum"," loan_amount"," loan_term"," cibil_score"," residential_assets_value"," commercial_assets_value"," luxury_assets_value"," bank_asset_value"]]
o=df[" loan_status"]
```

```python
lr.fit(i,o)


p=lr.predict([[3,0,1,3300000,11300000,20,559,4200000,2900000,11000000,1900000]])

if p>=0.5:
  print(" Aprroved")
else:
  print(" Not Approved")
```
