```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
```

```python
x = sb.load_dataset("tips")
x
```

```python
sb.scatterplot(x="total_bill",y="tip",data=x)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.barplot(x="total_bill",y="tip",data=x)
plt.title("bar Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.boxplot(x="total_bill",y="tip",data=x)
plt.title("box Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.lmplot(x="total_bill",y="tip",data=x)
plt.title("lmplot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.violinplot(x="total_bill",y="tip",data=x)
plt.title("violin")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.lineplot(x="total_bill",y="tip",data=x)
plt.title("Line Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.stripplot(x="total_bill",y="tip",data=x)
plt.title("Strip Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sb.jointplot(x="total_bill",y="tip",data=x)
plt.title("jointplot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
ir = sb.load_dataset("iris")
ir
```

```python
correlation_matrix = ir.drop('species', axis=1).corr()
sb.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
sb.pairplot(ir)
plt.show()
```

```python
sb.jointplot(ir)
plt.show()
```
